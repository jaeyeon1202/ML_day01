### loc를 이용한 조회

```python
# loc사용, 컬럼이 '경부선KTX'의 모든행
df.loc[:,['경부선KTX']]
```

```python
# 2011년 데이터만 선택, loc사용
df.loc[['2011',]]
```

```python
# 2013~2016년 행데이터 선택, loc사용
df.loc['2013':'2016',]
```

```python
# 2012~2014년까지 경부선KTX 컬럼만 조회, loc사용
df.loc['2012':'2014',['경부선KTX']]
```

```python
df['2012':'2014'][['경부선KTX']]
```

```python
# 2015~2017년까지 경부선KTX, 호남선KTX, 동해선KTX 컬럼 조회
df['2015':'2017'][['경부선KTX', '호남선KTX', '동해선KTX']]
```

```python
# isin사용
# 경부선등급이 'B', "C"인 데이터 조회
df[df['경부선등급'].isin(['B','C'])]
```

```python
# 위 조건을 만족하는 경부선KTX만 조회
df[df['경부선등급'].isin(['B','C'])][['경부선KTX']]
```
